Import packages

```python
import io
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
```

Conectar

```python
conn = sqlite3.connect('Banco.db')
cursor = conn.cursor()
```

### Consulta 1 

```python
# Deseja-se o nome, o registro e apresentação de todos os medicamentos comercializados em 2022, possuem restrição hospitalar e o regime de preço seja liberado.
# Versão 1
query = """
SELECT Registro, Nome, Apresentacao AS Apresentação
FROM MEDICAMENTO
WHERE Comercializacao2022 = 'Sim' AND RestricaoHospitalar = 'Sim' AND RegimeDePreco = 'Liberado'
"""

df = pd.read_sql_query(query, conn)
df
```

###  Consulta 2

```python
# Deseja-se o registro, o nome do produto e apresentação de todos os medicamentos que sejam do tipo fitoterápico, não possuam restrição hospitalar, sejam comercializados em 2022 e que possuam regime de preço regulado.
# Versão 1
query = """
SELECT
  Registro, Nome, Apresentacao AS Apresentação
FROM MEDICAMENTO
WHERE
  TipoDeProduto = 'Fitoterápico'
  AND RestricaoHospitalar = 'Não'
  AND Comercializacao2022 = 'Sim'
  AND RegimeDePreco = 'Regulado'
"""

df = pd.read_sql_query(query, conn)
df
```

 Consulta 3 

```python
# Deseja-se o nome dos medicamentos cuja finalidade seja de psicoestimulantes ou antiepilépticos e que tenham sido comercializados em 2022.
# Versão 1
query = """
SELECT DISTINCT Nome
FROM MEDICAMENTO AS M
INNER JOIN FINALIDADE AS F
  ON M.IdFi = F.IdFi
WHERE (DescricaoFi LIKE '%PSICOESTIMULANTE%' OR DescricaoFi LIKE "%ANTIEPILÉPTICO%") AND Comercializacao2022 = 'Sim'
"""

df = pd.read_sql_query(query, conn)
df
```

### Consulta 4

```python
# Deseja-se identificar o nome do(s) medicamentos(s), seu respectivo preço e apresentação, sendo tipo não genérico,
# cujo valor de comercialização máximo ao consumidor seja superior à R$ 100000,00 em Minas Gerais e ordenado de forma decrescente.
# Versão 1
query = """
SELECT DISTINCT
  M.Nome AS Remédio, ValorMGPMC, Apresentacao AS Apresentação
FROM
  MEDICAMENTO AS M
INNER JOIN
  PrecoMaxConsumidor AS P
  ON M.IdPMC = P.IdPMC
WHERE
  M.TipoDeProduto != 'Genérico'
  AND ValorMGPMC > 100000.00
  AND RestricaoHospitalar = 'Não'
ORDER BY ValorMGPMC DESC
"""

df = pd.read_sql_query(query, conn)
df
```

### Consulta 5

```python
# Deseja-se identificar o nome da(s) substância(s) que compõe o medicamento mais caro em termos de preço de comercialização máximo ao consumidor em Minas Gerais, seu preço nessas condições e nome.
# Versão 1
query = """
SELECT Nome AS Remédio, NomeSub AS Substância, ValorMGPMC
FROM PrecoMaxConsumidor AS P
INNER JOIN (Medicamento AS M
  INNER JOIN (Substancia AS S
    INNER JOIN ReSubMed AS R
      ON S.IdSub = R.IdSub)
    ON M.Registro = R.Registro)
  ON P.IdPMC = M.IdPMC
WHERE ValorMGPMC = (SELECT MAX(ValorMGPMC)
  FROM PrecoMaxConsumidor
  WHERE ValorMGPMC IS NOT NULL
)
"""

df = pd.read_sql_query(query, conn)
df
```

### Consulta 6

```python
# De acordo com Farmácia APP (2021), no primeiro semestre de 2021 as três categorias de medicamentos mais vendidas possuiam finalidade antidepressiva (primeiro), analgésica (segunda) e digestiva (terceira).
# Fonte: https://www.inovafarma.com.br/blog/medicamentos-mais-vendidos-em-farmacias/.
# Nesse sentido, deseja-se identificar o nome do laboratório fabricante, o nome do medicamento, sua respectiva finalidade, apresentação e valor máximo ao consumidor em Minas Gerais.
# Considere que o interesse está em considerar apenas os medicamentos com finalidade antidepressiva ou analgésica ou digestiva e que tenham valor máximo ao consumidor inferior à 10 reais, além de ser tarja preta.
# Versão 1
query = """
SELECT L.Nome AS Laboratório, M.Nome, DescricaoFi AS Finalidade, Apresentacao AS Apresentação, ValorMGPMC
FROM PrecoMaxConsumidor AS P
INNER JOIN (Tarja AS T
  INNER JOIN (FINALIDADE AS F
    INNER JOIN (MEDICAMENTO AS M
      INNER JOIN (Laboratorio AS L
        INNER JOIN Vinculo AS V
          ON L.CNPJ = V.CNPJ)
        ON M.Registro = V.Registro)
      ON F.IdFi = M.IdFi)
    ON M.IdTarja = T.IdTarja)
  ON P.IdPMC = M.IdPMC
WHERE (F.DescricaoFi LIKE '%ANTIDEPRESSIVO%'
OR F.DescricaoFi LIKE '%ANALG%'
OR F.DescricaoFi LIKE '%DIGEST%')
AND ValorMGPMC < 10.00
AND DescricaoTarja = 'Preta'
"""

df = pd.read_sql_query(query, conn)
df
```

### Consulta 7

```python
# Deseja-se identificar, por laboratório, seu nome, número de medicamentos que ele fabrica, o valor médio, máximo e mínimo dos preços não genéricos máximo ao consumidor praticados pela instituição em Minas Gerais.
# Considere apenas os laboratórios que produzem no mínimo 500 medicamentos e ordene em ordem crescente de quantidade produzida.
# Versão 1
query = """
SELECT
  L.Nome AS Laboratório, COUNT(*), AVG(ValorMGPMC), MAX(ValorMGPMC), MIN(ValorMGPMC)
FROM PrecoMaxConsumidor AS P
INNER JOIN (MEDICAMENTO AS M
  INNER JOIN (Laboratorio AS L
    INNER JOIN Vinculo AS V
      ON L.CNPJ = V.CNPJ)
    ON M.Registro = V.Registro)
  ON P.IdPMC = M.IdPMC
GROUP BY L.Nome
HAVING COUNT(*) >= 500
ORDER BY COUNT(*) ASC
"""

df = pd.read_sql_query(query, conn)
df
```
